# 検証①結局注目されている通信キャリアはどこか？


##背景
> 先月まで働いてきた業界。総務省から値下げ指示があったり、楽天が参入してきたりと何かと話題になる業界だが、世間の注目度はどうか。</br>後出しで最安値を出してきたauの社会的インパクトが気になる。

##仮説
> 最安値のインパクトは大きいのではないか。現状auの注目度・期待度は高い

##手法
> Google Trendsに置いてキーワード解析を行ってみる。

```
pytrend = TrendReq(hl='ja-jp')

keywords = ['ソフトバンク' , 'au', 'ドコモ']

pytrend.build_payload(kw_list=keywords, geo='JP', timeframe='2020-01-15 2021-01-15')
df_interest = pytrend.interest_over_time()
df_interest.plot(figsize=(12,4), marker='.')
plt.show()
```

# 検証②"楽天"で検索するときに人々が本当に調べたいものは何か

##背景
> 仮説①でモバイル事業について考察してみたが、そこで多角化経営している楽天の世間のイメージが気になった。

##仮説
> 既存EC事業のイメージがやはり強いのではないか。

##手法
> Google Trendsに置いてキーワード解析を行ってみる。WordCloudでビジュアル化してみる。

```
from collections import Counter
from wordcloud import WordCloud
```

```
# -*- coding: utf-8 -*-
from pytrends.request import TrendReq

# ---------- 設定 ----------

#抽出キーワード
KEYWORD = '楽天'
```

```
# キーワードリスト
kw_list = [KEYWORD]

# GoogleTrandsのリスト
google_trend_list = []
```

```
pytrends = TrendReq(hl='ja-JP', tz=360)
pytrends.build_payload(kw_list, cat=0, timeframe='today 12-m', geo='JP', gprop='')
trends = pytrends.related_queries()
print(trends)
```

```
trends_values = trends[KEYWORD]['top'].values.tolist()
print(trends_values)
```

```
for value in trends_values:
    item = str(value[0]).strip(KEYWORD)
    item = item.replace('　', '')
    item = item.replace(' ', '')
    google_trend_list.append(item)

print('google_trend_list = ' + str(google_trend_list))
```

```
# wordcloud用のフォントパス
WORDCLOUD_FONT_PATH = '/content/font/gomarice_mukasi_mukasi.ttf'

# wordcloud用の幅
WORDCLOUD_WIDTH = 1500

# wordcloud用の高さ
WORDCLOUD_HEIGHT = 1500

# wordcloud用の背景色
WORDCLOUD_BG_COLOR = 'white'

# wordcloud出力先
WORDCLOUD_OUTPUT_FILE = './wordcloud.png'
```

```
word_counter = Counter(google_trend_list)
wordcloud = WordCloud(
    font_path = WORDCLOUD_FONT_PATH,
    width = WORDCLOUD_WIDTH,
    height = WORDCLOUD_HEIGHT,
    background_color = WORDCLOUD_BG_COLOR).generate_from_frequencies(word_counter)
wordcloud.to_file(WORDCLOUD_OUTPUT_FILE)
```

*   分析詳細

*   考察
